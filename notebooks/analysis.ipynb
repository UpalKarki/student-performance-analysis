```python
import pandas as pd
import os
```

```python
BASE_DIR = os.path.dirname(os.getcwd())
data_path = os.path.join(BASE_DIR, "data", "students.csv")

df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["total"] = df["math"] + df["science"] + df["english"]
df["average"] = df["total"] / 3

df.head()
```

```python
df["result"] = df["average"].apply(lambda x: "Pass" if x >= 40 else "Fail")
df[["name", "average", "result"]]
```

```python
def grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 55:
        return "C"
    else:
        return "D"

df["grade"] = df["average"].apply(grade)
df[["name", "average", "grade"]]
```

```python
df["average"].hist()
```

```python
output_path = os.path.join(BASE_DIR, "data", "students_analysis.csv")
df.to_csv(output_path, index=False)
```
