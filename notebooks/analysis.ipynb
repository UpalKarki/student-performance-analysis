```python
# DAY 4 — EXPLORATORY DATA ANALYSIS (EDA)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
BASE_DIR = os.path.dirname(os.getcwd())
file_path = os.path.join(BASE_DIR, "data", "students.csv")

df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["total"] = df["math"] + df["science"] + df["english"]
df["average"] = df["total"] / 3
```

```python
plt.hist(df["average"])
plt.title("Distribution of Average Marks")
plt.xlabel("Average Marks")
plt.ylabel("Number of Students")
plt.show()
```

```python
# ### Observation:
# - Two students have average marks below 70, indicating lower performance
# - One student falls in the mid-performance range
# - Two students are high performers with averages above 85
```

```python
plt.scatter(df["attendance"], df["average"])
plt.xlabel("Attendance (%)")
plt.ylabel("Average Marks")
plt.title("Attendance vs Average Marks")
plt.show()
```

```python
# ### Attendance vs Performance Analysis

# The scatter plot shows a clear positive relationship between attendance and
# average marks. Students with higher attendance percentages tend to achieve
# higher average scores. This suggests that attendance may be an important
# factor influencing academic performance.
```

```python
df["result"] = df["average"].apply(lambda x: "Pass" if x >= 40 else "Fail")
df["result"].value_counts()
```

```python
df.groupby("result")["attendance"].mean()
```

```python
output_path = os.path.join(BASE_DIR, "data", "students_day4_analysis.csv")
df.to_csv(output_path, index=False)
```

```python
# DAY 5: RELATIONSHIPS, CORRELATION & CATEGORIZATION
```

```python
df[['attendance', 'average']].corr()
```

```python
#Anything compared with itself = 1.00
# Attendance vs Attendance → 1.00
# Average vs Average → 1.00
# 0.99 means Attendance and Average Marks move almost perfectly together
```

```python
def performance_category(avg):
    if avg < 70:
        return "Low"
    elif avg <= 85:
        return "Medium"
    else:
        return "High"

df['performance_level'] = df['average'].apply(performance_category)
df
```

```python
### DAY 5- Key Insights

# 1. Attendance and average marks show a strong positive correlation.
# 2. Students with attendance above 85% mostly fall in the High performance group.
# 3. Students below 70% attendance consistently show lower academic performance.
# 4. Improving attendance could potentially improve student outcomes.
```

```python
# DAY 7: Student Performance Classification & Insights
```

```python
df['average'] = df[['math', 'science', 'english']].mean(axis=1)
df[['student_id', 'average']]
```

```python
def classify_performance(avg):
    if avg >= 80:
        return "High"
    elif avg >= 65:
        return "Average"
    else:
        return "Low"

df['result'] = df['average'].apply(classify_performance)
df[['student_id', 'average', 'result']]
df['performance'] = df['average'].apply(classify_performance)
```

```python
# ### Day 9: Prediction Setup

# - Feature (X): Attendance
# - Target (y): Performance Category (High / Average / Low)
```

```python
# Can we predict student performance using attendance?
```

```python
def predict_performance(attendance):
    if attendance >= 85:
        return "High"
    elif attendance >= 70:
        return "Average"
    else:
        return "Low"
```

```python
df['attendance'] = df['attendance'].fillna(df['attendance'].mean())
```

```python
df['predicted_performance'] = df['attendance'].apply(predict_performance)
df[['name', 'attendance', 'performance', 'predicted_performance']]
```

```python
# ### Prediction Observations
# - Attendance-based rules provide a reasonable baseline prediction
# - Predictions closely align with actual performance for most students
# - Missing attendance values can affect predictions and must be handled beforehand
# - This rule-based approach helps understand feature importance before using ML
```
